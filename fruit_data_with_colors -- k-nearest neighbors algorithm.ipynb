```python
#Load the Data
Dataset = pd.read_table("fruit_data_with_colors.txt")
```

```python
#Create a dictionary and keep label and fruit name 
lookup_fruit_name = dict(zip(Dataset.fruit_label.unique(),Dataset.fruit_name.unique()))
lookup_fruit_name
```

# Split Data in train and test

```python
X = Dataset[["mass","width","height"]]
y = Dataset["fruit_label"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(X_train["mass"], X_train["width"], X_train["height"], c= y_train, marker="o", s=500)
ax.set_xlabel("width")
ax.set_ylabel("height")
ax.set_zlabel("color_score")
plt.show()
```

# Create Classifier Object

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
```

```python
#Train the classifier and fit the data
model.fit(X_train,y_train)
```

```python
#Accuracy
model.score(X_test, y_test)
```
